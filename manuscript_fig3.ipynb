```python
def rename_bands(vis_input, rename_mapping):
    for vis_entry in vis_input:
        for band in vis_entry['bands']:
            if band['band'][:-1] in rename_mapping:
                c_band = band['band'][:-1]
                c_orientation = band['band'][-1]
                band['band'] = rename_mapping[c_band] + c_orientation
            else:
                print(band['band'])
                raise RuntimeError('exists band in selected vis_input not in the mapping')
    return vis_input
```
